```python
import transformers
print(transformers.__version__)
```

```python
%%time
from transformers import pipeline

pipe = pipeline("text-to-speech")
```

```python
text = "Hello, how are you today?"
output = pipe(text)
```

```python
type(output)
```

```python
output
```

```python
output['sampling_rate']
```

```python
output['audio']
```

```python
from IPython.display import Audio

# Play the loaded audio
Audio(data=output['audio'],rate=output['sampling_rate'])
```

```python
from pydub import AudioSegment
```

```python
# If the voice is really slow and low, or high pitched and fast, your sampling rate is probably off!
audio_seg = AudioSegment(output['audio'].tobytes(), frame_rate=output['sampling_rate'],
                       sample_width = output['audio'].dtype.itemsize, channels = 1)
```

```python
audio_seg.export("output.mp3", format="mp3")# you can try wav
```
